```python
def minimum(*args):
    """Takes any number of arguments!"""
    m = args[0]
    for x in args[1:]:
        if x < m:
          m = x
    return m 
```

```python
minimum(6, 42)

data = [65, 42, 2, 8]
minimum(*data)
```

```python
def countdown():
    yield 3
    yield 2
    yield 1
    yield 'Blast off!'
```

```python
# generator
g = countdown()

next(g)
x = next(g)
print(x)
y, z = next(g), next(g)
print(z)
next(g)
```

```python
for t in countdown():
    if isinstance(t, int):
        message = "T-" + str(t)
    else:
        message = t
    print(message)
```

```python
# define a subgenerator
def yield_all(x): <1>
    for i in x:
        yield i

# palindrome using yield froms
def palindromize(x): <2>
    yield from yield_all(x) <3>
    yield from yield_all(x[::-1]) <4>

# the above is equivalent to this full expansion:
def palindromize_explicit(x): <5>
    for i in x: <6>
        yield i
    for i in x[::-1]: <7>
        yield i
```

```python
def null(f):
    """Always return None."""
    return

def identity(f):
    """Return the function."""
    return f

def self_referential(f):
    """Return the decorator."""
    return self_referential
```

```python
def plus1(f): <1>
    def wrapper(*args, **kwargs): <2>
        return f(*args, **kwargs) + 1 <3>
    return wrapper <4>
```

```python
@plus1
def power(base, x):
    return base**x

power(4, 2)
```

```python
@plus1
@identity
@plus1
@plus1
def root(x):
    return x**0.5

root(4)
```

```python
def plus_n(n): <1>
    def dec(f): <2>
        def wrapper(*args, **kwargs): <3>
            return f(*args, **kwargs) + n <4>
        return wrapper <5>
    return dec <6>
```

```python
@plus_n(6)
def root(x):
    return x**0.5

root(4)
```
